```python
df[(df.KECEPATAN == 100) & (df.KATEGORI_PELANGGAN == "RUMAH KOST")][["KECEPATAN","KATEGORI_PELANGGAN","PAKET", "DATEL"]]
```

```python
df_jum_speed=df[df.KATEGORI_PELANGGAN == "RUMAH KOST"].groupby(["KECEPATAN"])[["KECEPATAN"]].count().rename(
    columns={"KECEPATAN":"COUNT"}).reset_index()
df_jum_speed.head()
```

```python
df_jum_speed=df[df.KATEGORI_PELANGGAN == "PRIME CLUSTER"].groupby(["KECEPATAN"])[["KECEPATAN"]].count().rename(
    columns={"KECEPATAN":"COUNT"}).reset_index()
df_jum_speed.head()
```

```python
df_jum_speed=df[df.KATEGORI_PELANGGAN == "APARTMENT"].groupby(["KECEPATAN", "PAKET"])[["KECEPATAN"]].count().rename(
    columns={"KECEPATAN":"COUNT"})
df_jum_speed.sort_index()
```

```python
df_agg = df.groupby(["KATEGORI_PELANGGAN", "PAKET"])["KECEPATAN"].mean().reset_index()
df_agg = df_agg.rename(index=str, columns={"KECEPATAN": "RATA-RATA KECEPATAN YANG DIPILIH"})
df_agg
```

```python
df = pd.merge(df, df_agg, how='left',on=["KATEGORI_PELANGGAN", "PAKET"])
df
```
